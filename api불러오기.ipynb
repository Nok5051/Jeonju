```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

# 1. 가로등 데이터

```python
url_street = 'http://openapi.jeonju.go.kr/rest/streetlamp/getStreetlamp?authApiKey='
open_api_key = '30NUDEQ1MBHnAJ6uehvw8Bman579Pq9xYEkMgIYWL1cYKLaNtfP1I4eLSZVmTnDz2juz9RaCLQKoUbWRBYcIYQ%3D%3D'
pagesize = '&pageSize=1000'
```

```python
total_counts = 14935
page_size = 1000
pages = int(total_counts / page_size) + 1
streetlamp = []

for i in range(pages):
    print('getting data from page %s of page %s'%(i+1,pages))
    open_url = url_street + open_api_key + '&startPage=' + str(i+1) + pagesize
    
    req = requests.get(open_url)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    soup_list = soup.find_all('list')
    
    var_lists = ['basedate','boxnm','lampcnt','polenum','posx','posy','seq']
    
    for n in range(len(soup_list)):
        a=[]
        for var in var_lists:
            found_var = soup_list[n].find_all(var)
            found_var = [x.text for x in found_var]
            a += found_var
            
        streetlamp.append(a)
```

```python
df1 = pd.DataFrame(streetlamp)
df1
```
